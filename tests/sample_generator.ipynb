# cdc_loader — exploratory tests

```python
import subprocess, sys
from pathlib import Path

# Resolve the repo root regardless of the notebook's working directory
repo_root = Path.cwd()
while not (repo_root / "scripts" / "generate_sample_data.py").exists():
    if repo_root.parent == repo_root:
        raise FileNotFoundError("Could not locate repo root (scripts/generate_sample_data.py not found)")
    repo_root = repo_root.parent

result = subprocess.run([
    sys.executable,
    str(repo_root / "scripts" / "generate_sample_data.py"),
    "--rows", "50000",
    "--days", "30",
    "--merchants", "100",
    "--seed", "42",
    "--out", str(repo_root / "docs" / "sample-data"),
], cwd=repo_root, capture_output=True, text=True)

print(result.stdout)
print(result.stderr)
result.check_returncode()
```

```python
import subprocess, sys

result = subprocess.run([
    sys.executable, "-m", "pytest",
    str(repo_root / "tests" / "test_generate_sample_data.py"),
    "-v",
], cwd=repo_root, capture_output=True, text=True)

print(result.stdout)
print(result.stderr)
result.check_returncode()
```
