# cdc_loader — exploratory tests

```python
import sys
from pathlib import Path

PROJECT_ROOT = Path(r"C:\Users\Leal\data-engineer-challenge")
SCRIPT_PATH = PROJECT_ROOT / "scripts" / "generate_sample_data.py"

sys.path.insert(0, str(PROJECT_ROOT))
import subprocess
from pathlib import Path
import pyarrow.parquet as pq
import pandas as pd
```

```python
# ── Configuração do teste ──────────────────────────────────────────────────
OUT_DIR = Path(r"C:\Users\Leal\data-engineer-challenge\tmp_sample_data_test")
ROWS = 1000
DAYS = 5
MERCHANTS = 10
SEED = 42

# ── Limpa pasta anterior (o gerador não faz isso sozinho) ─────────────────
import shutil
if OUT_DIR.exists():
    shutil.rmtree(OUT_DIR)
OUT_DIR.mkdir(parents=True)

# ── Roda o gerador como subprocess ─────────────────────────────────────────
result = subprocess.run(
    [
        sys.executable, str(SCRIPT_PATH),
        "--rows", str(ROWS),
        "--days", str(DAYS),
        "--merchants", str(MERCHANTS),
        "--seed", str(SEED),
        "--out", str(OUT_DIR),
    ],
    capture_output=True, text=True
)

print("STDOUT:")
print(result.stdout)
print("STDERR:")
print(result.stderr)
print("Return code:", result.returncode)
```

```python
# ── Confere o que foi gerado ────────────────────────────────────────────────
print("Arquivos na raiz de OUT_DIR:")
for f in sorted(OUT_DIR.glob("*")):
    print(" ", f.name)

print("\nArquivos em paysettler/:")
paysettler_dir = OUT_DIR / "paysettler"
if paysettler_dir.exists():
    for f in sorted(paysettler_dir.glob("*.csv")):
        print(" ", f.name)
```
